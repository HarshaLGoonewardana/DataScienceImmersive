## How to make XKCD plots in Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3]
y = [2,3,3]
plt.plot(x,y)
plt.title('This is not an XKCD plot:');
```

```python
x = [1,2,3]
y = [2,3,3]
plt.xkcd() # This is all you need!
plt.plot(x,y)
plt.title('This is an XKCD plot,\n but the font isn\'t correct:');
```

```python
# Import font_manager utility
from matplotlib import font_manager

# List system fonts:
font_list = font_manager.findSystemFonts(fontpaths=None, fontext='ttf')

# Calibri fonts:
[font_list[i] for i in range(len(font_list)) if ['calibri' in font for font in font_list][i]]
```

In order to get the full XKCD effect, you need the Humor-Sans font found [here](http://antiyawn.com/uploads/humorsans.html). Download it and then put it in your fonts folder. (The folder in which calibri fonts can be found...)

```python
# Success!
[font_list[i] for i in range(len(font_list)) if ['humor' in font for font in font_list][i]]
```

```python
# Rebuild the list of available fonts from system fonts
font_manager._rebuild()
```
