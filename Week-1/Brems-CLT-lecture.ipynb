```python
def sampler(population, n, k):
    poll_results = []
    for i in range(k):
        p_hat = np.mean(np.random.choice(population, size = n, replace = False))
        poll_results.append(p_hat)
    return poll_results
```

```python
polls_size10_1000polls = sampler(popn, 10, 1000)
```

```python
sns.distplot(polls_size10_1000polls);
```

```python
polls_size20_1000polls = sampler(popn, 20, 1000)
sns.distplot(polls_size20_1000polls);
```

```python
polls_size30_1000polls = sampler(popn, 30, 1000)
sns.distplot(polls_size30_1000polls);
```
